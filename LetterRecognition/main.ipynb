```python
import torch.nn as nn

class CharacterNet(nn.Module):

    def __init__(self, in_channels: 1) -> None:
        super().__init__()
        self.conv1 = nn.Conv2d(in_channels = in_channels, out_channels = 32, kernel_size = 3, stride = 1, padding = 1)  # in: 1x28x28, out: 32x28x28
        self.mp1 = nn.MaxPool2d(kernel_size = 2, stride = 2) # in: 32x28x28, out: 32x14x14
        self.conv2 = nn.Conv2d(in_channels = 32, out_channels = 64, kernel_size = 3, stride = 1, padding = 1) # in: 32x14x14, out: 64x14x14
        self.mp2 = nn.MaxPool2d(kernel_size = 2, stride = 2) # in: 64x14x14, out: 64x7x7
        self.lin1 = nn.Linear(in_features = 3136, out_features = 128) # in: 3136, out: 128
        self.lin2 = nn.Linear(in_features = 128, out_features = 26) # in: 128, out: 26

    def forward(self, x):
        x = self.conv1(x)
        x = self.mp1(x)
        x = self.conv2(x)
        x = self.mp2(x)
        torch.flatten(x, start_dim = 1)
        x = x.view(-1, 3136) # Return a new tensor with the same data as the self tensor but of a different shape.
        x = self.lin1(x)
        x = self.lin2(x)
        return x
```

```python

char_net = CharacterNet(in_channels=1)

```
